# **Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating Toy Dataset
data = {'x1':[0,0,1,1], 'x2':[0,1,0,1], 'Y' :[0,0,0,1]}
data = pd.DataFrame(data)
data
```

```python
import numpy as np
```

```python
# Initialising W values:
W = np.array([0.1,0.1,0.1])
W = W.reshape(3,1)
```

```python
def SGD(eta,i,sig,W,xi):
    W = W + eta*(data['Y'][i] - sig)*xi
    return W
```

```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
wt = []
```

```python
for k in range(6):
  for i in range(len(data)):
      t = W.transpose()
      xi = np.array([1,data['x1'][i],data['x2'][i]])
      xi = xi.reshape(3,1)
      wtx = np.dot(t,xi)
      wt.append(wtx[0][0])
      sig = sigmoid(wtx)
      W = SGD(0.1,i,sig,W,xi)
    
print(W)    
```

```python
xl = np.linspace(-5,5,num=10)
x2 = np.linspace(-4,4,num=10)
pred = []
for i in range(len(xl)):
    pred.append(W[0][0] + W[1][0]*xl[i] + W[2][0]*x2[i])
```

## Plotting

```python
plt.scatter(data['x1'],data['x2'],marker = 'o',color = 'red')
plt.plot(pred,color ='blue')
```

```python
from mpl_toolkits.mplot3d import Axes3D

xx, yy = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
z = (W[1] * xx + W[2] * yy + W[0])
 
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(xx, yy, z, alpha=0.2)
 
ax = plt.gca()
ax.scatter3D([0, 0, 1, 1], [0, 1, 0, 1], [0, 0, 0, 1])
plt.show()
```
